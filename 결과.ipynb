#**DSANet**

```python
import matplotlib.pyplot as plt

# 모델 평가
model.eval()
with torch.no_grad():
    predicted = model(train_data).numpy()  # 예측 결과 [batch_size, num_nodes, value_dim]

# 실제 데이터
actual = train_data[:, 0, 0].numpy()  # [batch_size]

# 예측 결과의 차원을 줄임
# 예: value_dim에 대해 평균을 내어 [batch_size, num_nodes] -> [batch_size]로 축소
predicted = predicted.mean(axis=(1, 2))  # 모든 차원을 평균하여 [batch_size]로 만듦

plt.figure(figsize=(12, 8))

# 실제 데이터 시각화
plt.plot(actual, label='Actual', color='blue')

# 예측 데이터 시각화
plt.plot(predicted, label='Predicted', color='red')

plt.title('Actual vs Predicted Speeds (First Route)')
plt.xlabel('Time Step')
plt.ylabel('Speed')
plt.legend()
plt.grid(True)
plt.show()
```
